|5.0|1.00|
|6.0|0.83|
|7.0|0.71|
|8.0|0.62|

For your plot of voltage ($y$-axis) versus distance ($x$-axis): Have a solid, thick green line with no marker (use keyword input parameters), start both axes at 0.0 and end them on their maximum values, place tick marks (with labels) on the $x$-axis at the even values of distance, label your axes and title the plot appropriately, and put your name on the plot using the `text` function. Section 5.2.2 in the textbook describes ways to customize a line plot. As a hint, Chapter 6 in the textbook discusses `max`. Remember to include an $x$- and $y$-location for the text on the plot.

**Try This! for Physics 5-4** Several students are running track. They are trying to maintain a constant speed as part of their conditioning for running long distances. Their (fictitious) masses and speeds are as follows.

|Mass $(\mathrm{kg})$|Speed $(\mathrm{m/s})$|
| :--: | :--: |
|51.1|3.83|
|62.7|3.58|
|49.8|3.48|
|65.2|3.88|
|66.0|4.06|

What is the average kinetic energy for the group? Recall that kinetic energy $(\mathrm{KE})$ is

$$\mathrm{KE} = \frac{1}{2}mv^2$$ 

Use array syntax to perform your calculation. Section 5.2.6 in the textbook introduces array calculations.

**Solution** 

````python
import numpy as np

mass = np.array([51.1, 62.7, 49.8, 65.2, 66.0])
speed = np.array([3.83, 3.58, 3.48, 3.88, 4.06])
kinetic_energy = 0.5 * mass * speed ** 2
np.mean(kinetic_energy)
````

**Try This! for Physics 5-6** In the lab, you collect (fictitious) data on the time and momentum of a cart.

|Time $(\mathrm{s})$|Momentum $(\mathrm{kg}\cdot\mathrm{m/s})$|
| :--: | :--: |
|0.0|0.0|
|0.5|1.5|
|1.0|1.8|
|1.5|1.9|
|2.0|2.2|
|2.5|1.7|
|3.0|3.0|
|3.5|4.5|
|4.0|2.4|
|4.5|1.3|
|5.0|0.0|

Approximate the net force the cart experiences for each neighboring time interval, e.g., from $0.0\ \mathrm{s}$ to $0.5\ \mathrm{s}$, $0.5\ \mathrm{s}$ to $1.0\ \mathrm{s}$, etc., by calculating the change in momentum ($dp$) for each neighboring time interval, the change in time ($dt$) for each neighboring time interval, and dividing the two, since

$$F_{net} = \frac{dp}{dt}$$ 

Use array syntax. Section 5.2.6 in the textbook introduces array calculations. Try This! 5-5 in the textbook shows how to use array syntax to calculate such differences.

**Solution**

````python
import numpy as np

time = np.array([0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0])
momentum = np.array([0.0, 1.5, 1.8, 1.9, 2.2, 1.7, 3.0, 4.5, 2.4, 1.3, 0.0])
momentum_change = momentum[1:] - momentum[:-1]
time_change = time[1:] - time[:-1]
force = momentum_change / time_change
force
````

Note that `momentum[1:]` slices the `momentum` array from the second element (1.5) to the end such that it omits the first element (0.0), while `momentum[:-1]` slices the `momentum` array from the first element to the second-to-last element such that it omits the last element. This allows you to take the difference between each successive momentum element for each 0.5 s time difference. A similar procedure is carried out on the `time` array.

**Try This! for Physics 5-7** Your task is to lift a very heavy object using a lever arm: a board that is wedged under the mass and set on a pivot point (see the image below). Define a function that finds the minimum force, $F$, that needs to be applied straight down to lift the mass, $M$. The function should have positional input parameters $M$ and $d$ (the distance from the pivot point to where the force, $F$, is applied).  The function should have optional, keyword input parameters for $x$ (the distance from the mass to the pivot point) and $g$ (the acceleration due to gravity).

<img src="Ch5LeverArm.jpg" width="300" align="center">

Set the default values of the keyword input parameaters to $1.0\ \mathrm{m}$ for $x$ and $9.8\ \mathrm{m/s^{2}}$ for $g$. Define the function in one cell and use it in a second cell. Assume that the mass of the lever arm is negligible and that the mass of the object sits at the end of the lever arm such that $x$ represents the distance from the pivot point to the left end of the board.

Recall that this is a static equilibrium problem since the **minimum** force is what is being calculated, i.e., the force needed right before angular acceleration occurs. With the pivot point as the origin, the force applied by the person at a displacement $d$ to the right of the pivot point provides a torque that points into the page, and that torque is balanced by the torque that points out of the page that is provided by the force ($Mg$ where $g=9.8\ \mathrm{m/s^{2}}$) applied by the mass at a displacement of $x$ to the left of the pivot point. We can then use the equation for torque

$$\vec{\tau} = \vec{r} \times \vec{F}$$ 

and the sum of torques equation

$$\Sigma \vec{\tau} = I\vec{\alpha}$$ 

which will equal $0\ \mathrm{N\cdot m}$ (since the angular acceleration, $\vec{\alpha}$, is $0\ \mathrm{rad/s^{2}}$) in order to find $F$.

For a cross-product, the magnitude can be found by multiplying the vectors by the sine of the angle between them: 

$$\tau = rF\sin(\theta)$$ 

Therefore

$$dF\sin(\beta) = xMg\sin(\phi)$$ 

and solving for $F$ gives

$$F = \frac{xMg\sin(\phi)}{d\sin(\beta)}$$ 

Because $\phi = \pi - \beta,$ $\sin(\phi) = \sin(\beta),$ then 

$$F = \frac{x}{d}Mg$$

which is the minimum force, $F$, that needs to be applied straight down to lift the mass. Section 5.2.1 in the textbook describes using keyword input parameters for optional input.

**Solution** 

````python
import numpy as np

def lever_arm_force(mass, force_displacement, mass_displacement = 1.0, grav_accel = 9.8):
    return (mass_displacement / force_displacement) * mass * grav_accel
````

````python
lever_arm_force(50, 1.5, mass_displacement = 1.5)
````

To test the function, the `force_displacement` has been set to the same length as the `mass_displacement` to make sure that it would take the equivalent amount of force as $Mg$.

Use array syntax to calculate the angular velocity for each acceleration, given that the relationship between centripetal acceleration $(a_{c}, \mathrm{in\ m/s^2})$, radius $(r, \mathrm{in\ m})$, and angular velocity $(\omega, \mathrm{in\ rad/s})$ is

$$a_{c} = r \omega^{2}$$ 

and therefore

$$\omega = \sqrt{\frac{a_{c}}{r}} = \left(\frac{a_{c}}{r}\right)^{\frac{1}{2}}$$ 

Make a plot with $\omega$ on the $x$-axis and $a_{c}$ on the $y$-axis, make the marker a diamond, and make the line connecting the points a green, dashed line. Label all axes appropriately. **Try This! for Physics 5-9**  Pretend that you are observing three moving carts over the course of a four-minute period. At given times during the period, you observe the speed of each cart in meters per second $(\mathrm{m/s})$. You obtain the following ficticious data.

| Time $(\mathrm{s})$ | Speed $(30\ \mathrm{g})$ | Speed $(40\ \mathrm{g})$ | Speed $(70\ \mathrm{g})$ |
|:--------:|:------------:|:------------:|:------------:|
|  60      |    1.9       |     1.8      |     1.1      |
|  120     |    2.1       |     1.4      |     1.3      |
|  180     |    2.3       |     2.0      |     1.3      |
|  240     |    1.8       |     1.8      |     1.2      |

Recalling that the kinetic energy (KE) is given by

$$ \mathrm{KE} = \frac{1}{2} m v^2 $$

where $m$ is the mass in kilograms $\mathrm{kg}$ and $v$ is the velocity in meters per second $\mathrm{m/s}$, create three plots of kinetic energy ($y$-axis) versus time ($x$-axis), where each plot is a separate figure showing the variation of kinetic energy with time for the 30 g cart, the 40 g cart, and the 70 g cart. (Note: You have to convert from grams to kilograms for the mass of the carts. Use array syntax to calculate kinetic energy.) Make sure each plot is correctly labeled and uses different colors, markers, and linetypes. Make each marker 6 point in size.

Section 5.2.2 in the textbook describes ways to customize a line plot.  Section 5.2.3 describes handling multiple figures.

**Solution**

````python
import numpy as np
import matplotlib.pyplot as plt

time = np.array([60, 120, 180, 240])
velocity_1 = np.array([1.9, 2.1, 2.3, 1.8])
velocity_2 = np.array([1.8, 1.4, 2.0, 1.8])
velocity_3 = np.array([1.1, 1.3, 1.3, 1.2])

kinetic_energy_1 = 0.5 * (30.0 / 1000.0) * (velocity_1**2)
kinetic_energy_2 = 0.5 * (40.0 / 1000.0) * (velocity_2**2)
kinetic_energy_3 = 0.5 * (70.0 / 1000.0) * (velocity_3**2)

marker_size_in_pts = 6.0

plt.figure('Cart 1')
plt.plot(time, kinetic_energy_1, 'r-o', markersize = marker_size_in_pts)
plt.xlabel('Time (s)')
plt.ylabel('Kinetic Energy (J)')
plt.title('Cart 1: 30 g')

plt.figure('Cart 2')
plt.plot(time, kinetic_energy_2, 'b-.D', markersize = marker_size_in_pts)
plt.xlabel('Time (s)')
plt.ylabel('Kinetic Energy (J)')
plt.title('Cart 2: 40 g')

plt.figure('Cart 3')
plt.plot(time, kinetic_energy_3, 'g--*', markersize = marker_size_in_pts)
plt.xlabel('Time (s)')
plt.ylabel('Kinetic Energy (J)')
plt.title('Cart 3: 70 g')

plt.show()
````

For a better solution, you can save the masses to variables and then use those variables both for calculating the kinetic energy as well as for the titling of the graphs.  This enables you to define each mass in one place, rather than in two places as in the solution above (i.e., once in the expression to calculate kinetic energy and once in the title).  When you define the same value in multiple places, rather than saving the value as a variable and then using the variable wherever you want to use the value, the repetition makes the code more prone to errors.
